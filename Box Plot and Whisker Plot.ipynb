# Box Plot

```python
#upper and lower part of box plot is whiskey plot
# o in outer part represents that the value is out from the range
#yellow line represents mediaan
import matplotlib.pyplot as plt
x=[10,20,30,40,50,60,70,120]
x1=[10,20,40,50,30,90]
y=[x,x1]
#plt.boxplot(x,notch=True)
#plt.boxplot(x,vert=False)
#plt.boxplot(x,widths=0.8)
#plt.boxplot(x,labels=["python"],patch_artist=True,showmeans=True)
#plt.boxplot(x,labels=["python"],showmeans=True,whis=2)
plt.boxplot(y,labels=["python","c++"],capprops=dict(color="g"),whiskerprops=dict(color="r"),flierprops=dict(markeredgecolor="b"))
plt.show()
```
