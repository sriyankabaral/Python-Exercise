# Scatter Plot

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7]
y=[2,3,1,5,3,7,6]
plt.scatter(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,5,3,7,6]
plt.scatter(day,no, color="r")
plt.title("Scatter plot",fontsize=15)
plt.xlabel("Day",fontsize=12)
plt.ylabel("Number")
plt.show()
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,5,6,7,6]
colors=["r","g","y","b","m","y","b"]
sizes=[100,200,400,700,222,459,664]
plt.scatter(day,no, c=colors,s=sizes)
plt.title("Scatter plot",fontsize=15)
plt.xlabel("Day",fontsize=12)
plt.ylabel("Number")
plt.show()
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,5,6,7,6]
colors=["r","g","y","b","m","y","b"]
sizes=[100,200,400,700,222,459,664]
plt.scatter(day,no, c=colors,s=sizes, alpha=0.4)
plt.title("Scatter plot",fontsize=15)
plt.xlabel("Day",fontsize=12)
plt.ylabel("Number")
plt.show()
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,5,6,7,6]
colors=["r","g","y","b","m","y","b"]
sizes=[100,200,400,700,222,459,664]
plt.scatter(day,no, c=colors,s=sizes, alpha=0.4,marker="*")
plt.title("Scatter plot",fontsize=15)
plt.xlabel("Day",fontsize=12)
plt.ylabel("Number")
plt.show()
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,5,6,7,6]
colors=["r","g","y","b","m","y","b"]
sizes=[100,200,400,700,222,459,664]
plt.scatter(day,no, c=colors,s=sizes, alpha=0.4,marker="*",edgecolor="black",linewidth=2)
plt.title("Scatter plot",fontsize=15)
plt.xlabel("Day",fontsize=12)
plt.ylabel("Number")
plt.show()
```

```python
# color code
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,5,6,7,6]
# colors=["r","g","y","b","m","y","b"]
colors=[10,20,34,45,67,66,77]
sizes=[100,200,400,700,222,459,664]

# plt.scatter(day,no, c=colors,s=sizes,cmap="BrBG")
plt.scatter(day,no, c=colors,s=sizes,cmap="viridis",alpha=0.6)

t=plt.colorbar()
t.set_label("ColorBar",fontsize=15)

plt.title("Scatter plot",fontsize=15)
plt.xlabel("Day",fontsize=12)
plt.ylabel("Number")
plt.show()
```
