# PieChartPlot

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["c","c++","java","python"]
ex=[0.3,0.0,0.0,0.0]
c=["r","b","g","y"]
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.2f%%",shadow=True,radius=1.3,labeldistance=1.1,startangle=180,textprops={"fontsize":15},counterclock=False,wedgeprops={"linewidth":5,"edgecolor":"m"},
       center=(3,3),rotatelabels=False)
plt.title("piechart")
plt.legend(loc=2)
plt.show()
```
