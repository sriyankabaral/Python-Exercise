# Bar Plot

```python
import matplotlib.pyplot as  plt
x=["python","c","c++","java"]
y=[85,70,60,82]
z=[20,30,40,50]
plt.xlabel("language",fontsize=10)
plt.ylabel("number",fontsize=10)
plt.title("sriyanka",fontsize=10)
c=["y","g","b","r"]

#plt.bar(x,y,width=0.5,color=c,align="edge",edgecolor="g",linewidth=2,linestyle=":")
plt.bar(x,y,width=0.5,color="r",align="edge",edgecolor="g",label="popularity")
plt.bar(x,z,width=0.5,color="b",align="edge",edgecolor="g",label="popularity1")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as  plt
import numpy as np

x=["python","c","c++","java"]
y=[85,70,60,82]
z=[20,30,40,50]

p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("language",fontsize=10)
plt.ylabel("number",fontsize=10)
plt.title("sriyanka",fontsize=10)
width=0.1

plt.bar(p,y,width,color="r",label="popularity")
plt.bar(p1,z,width,color="b",label="popularity1")

plt.xticks(p+width/2,x,rotation=12)
plt.legend()
plt.show()
```
